# If Elif Else Statements

## Branches to the condition

```python
# Calculate the letter grade score with IF ... ELIF
#not checking for boundaries unless specify
#runs sequentially in data flow/linear
#if condition in number one met will not even run a later condition
#winner takes all
# want adjacent if statements if want all to be evaluated every single time vs with if elif where only one line is taken til conditions met

score = 1000

if score >= 90:
    print("is it an A")
    letter = 'A'
elif score >= 80:
    print("is it a b?")
    letter = 'B'
elif score >= 70:
    letter = 'C'
elif score >= 60:
    letter = 'D'
elif score >= 1000:
    letter = 'Cheating'
else:
    letter = 'F'
    
print("The score is: -> ", letter)
```

```python
age = 99

if age>30:
    print("You're over thirty")
if age<30:
    print("You're under thirty")

# in this case two adjacent if statements
#both run independently

#coupled statements - if elif or nested if else statments, will only run until hit the line that matches your condition
```

```python
#  The for-loop can be used to iterate over characters of a string
string1 = '123456789'

sum = 0
for char in string1:
    print(f"Considering {char}...")
    sum += int(char)
    print(f"Sum is now [sum].")
print(sum)
```
