```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
```

```python
base = np.arange(0, 5, 0.5)
```

```python
print(base)
print(len(base))
```

```python
# taking the base creation and splitting into x and y variables

x2 = base[:6]
y2 = base[:6]**2

x3 = base[5:]
y3 = base[5:]**2

# **(whatever number) is an exponential, 10**2 is 10 raised to the second power so 10*10 = 100

print(x2)
print(y2)

print(x3)
print(y3)
```

```python
plt.plot(x,y, color = "salmon", linewidth = "1", linestyle = "-.", marker = "*", markersize = 10)
plt.show()
# x is your x coordinate, your y is your y coordinate, and your marker shows you the specific points that made the line graph
#we can change our line style, default is a complete line, can cusotmize
#can change line color, can use hex code/value or color name
#can change marker size
#can change width of the line
##.show prints the graph like print does for tables and strings and such
```

```python
plt.plot(x,y, "b^--")
##shorthand
##set line to b for blue
#set marker for that symbol of the carrot
#set the line style to dotted line
plt.show()
```

```python
plt.plot(x,y, "b^-", label = "2x", markersize = 10, linewidth = 1)

plt.plot(x2,y2, "r^-", label = "x*2", markersize = 10, linewidth = 1)

plt.plot(x3,y3, "g^--", label = "forecast", markersize = 10, linewidth = 1)

plt.title("Our First Graph!", fontdict = {"fontname" : "Comic Sans MS", "fontsize" : 20})

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

#overrides default tick mark values on the axes
plt.yticks([0,5,10,15,20])
plt.xticks([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])

plt.legend()
plt.show()
```
